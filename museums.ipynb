```python
import folium
from folium.plugins import FastMarkerCluster
import csv
import json
```

```python
coors = []
with open('museum_coor.csv', 'r') as f:
    t = csv.reader(f,delimiter=',', dialect=csv.excel, )
    for all in t:
        if all:
            coors.append(all)
new_coors = []
for all in coors:
    new_coors.append([all[1], all[0]])
```

```python
with open('Regions.json', 'r', encoding="utf8") as f:
    i = json.load(f)
```

```python
new_json = {}
new_json['type'] = 'FeatureCollection'
new_json['features'] = []
count = 0
for all in i:
    count += 1
    a = {}
    a['type'] = 'Feature'
    a['id'] = str(count)
    a['properties'] = {}
    a['properties']['name'] = all
    a['geometries'] = {}
    a['geometries']['type'] = 'Polygon'
    coor = []
    #a['geometry']['coordinates'] = io[all]['0']
    for k, v in i[all].items():
        coor.append(i[all][k])
    a['geometries']['coordinates'] = coor
    new_json['features'].append(a)
```

```python
m = folium.Map(location=[66, 100], zoom_start=3)

for all in new_json['features']:
    folium.Polygon(all['geometries']['coordinates'], color = 'green', fill = True, fillColor = 'blue', weight=0.5, opacity=1).add_to(m)

FastMarkerCluster(data=new_coors).add_to(m)

folium.LayerControl().add_to(m)

m.save('museums_map.html')
m
```
